```python
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
from dotenv import load_dotenv
import os

load_dotenv()

csv_path_real = os.getenv("REAL_DATA_PATH")

RHP = pd.read_csv(csv_path_real)

if csv_path_real is None:
    raise ValueError("DATA_CSV_PATH not set in .env file")
```

```python
model = ARIMA(RHP, order = (1,2,1))

model_fit = model.fit()

forecast = model_fit.forcast(steps = 5)
```

```python
csv_path_predicted = os.getenv("PREDICTED_DATA_PATH")

df_forcast = pd.dataframe({"Predicted Price" : forcast})

df_forcast.to_csv(csv_path_predicted, index = False)
```
