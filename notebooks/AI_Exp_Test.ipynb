```python
from ollama import chat
import pandas as pd
from dotenv import load_dotenv
import os

MODEL = "guanaco.3b"

load_dotenv()

csv_path_real = os.getenv("REAL_DATA_PATH")
csv_path_predicted = os.getenv("PREDICTED_DATA_PATH")

RHP = pd.read_csv(csv_path_real)

if csv_path_real is None:
    raise ValueError("DATA_CSV_PATH not set in .env file")

PHP = pd.read_csv(csv_path_predicted)

if csv_path_predicted is None:
    raise ValueError("DATA_CSV_PATH not set in .env file")

combined_df = pd.concat([RHP, PHP], ignore_index = True)


data_from = 2006
present = 2026
prediction_range = 5
```

```python
def explain_graph(params: dict) -> str:
    prompt = f"""
You are explaining a graph to a non-technical user.

Graph info (precomputed):
{params}

Write:
1) One short plain-English paragraph explaining what the chart shows.
2) 3 bullet key takeaways.
3) One limitation/uncertainty sentence.
Keep it friendly and clear.
""".strip()

r = chat(model = MODEL, message = [{"role" : "user", "content": prompt}])
return r["message"]["content"]

params = {
    "title" : "Uk Median House Prices"
    "range" : ((present + prediction_range) - data_from)
    "start" : combined_df[0]
    "end" : combined_df[-1]
}
```
